```python
import numpy as np
import pandas as pd
from pathlib import Path

import warnings
warnings.filterwarnings('ignore')
```

```python
scrdir = Path("./")
file1 = scrdir / "wdc.apf"
file2 = scrdir / "dug.apf"
```

```python
columns = [
    "event name",
    "station name",
    "component azimuth",
    "component incident angle / pick id",
    "year of pick / julian day",
    "hours",
    "minutes",
    "seconds",
    "amplitude of pick",
    "source of pick ['A'--APK; 'M'--PPK]",
    "type of file"
]

wrt_cols = [
    "station name",
    "year of pick / julian day",
    "hours",
    "minutes",
    "seconds"
]
```

```python
df1 = pd.read_csv(file1, delimiter='\s+', index_col=False, header=None, names=columns)
df1.head()
```

```python
outfile1 = scrdir / "wdc.pktimes"
df1.to_csv(outfile1, sep='\t', columns=wrt_cols, index=False, header=False)
```

```python
df2 = pd.read_csv(file2, delimiter='\s+', index_col=False, header=None, names=columns)
df2.head()
```

```python
outfile2 = scrdir / "dug.pktimes"
df2.to_csv(outfile2, sep='\t', columns=wrt_cols, index=False, header=False)
```
