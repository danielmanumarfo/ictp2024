```python
# set station epicentral distances
s1d = 10112.5
s2d = 10934.1
deltad = s2d - s1d
print(deltad)
```

```python
# Read first file 'wdc.pktimes'
st1 = []
with open(scrdir/'wdc.pktimes', 'r') as file1:
    for line in file1:
        hr, mm, sec = map(float, line.split()[2:5])
        atime = hr*3600+mm*60+sec
        ttime = atime - otime
        st1.append(ttime)
ns1 = len(st1)
```

```python
# Compute t1 and u1
t1, u1 = [], []
for i in range(ns1-2):
    t = st1[i+2] - st1[i]
    v = s1d / st1[i+1]
    t1.append(t); u1.append(v)

plt.plot(t1, u1)
plt.grid()
plt.show()
```

```python
# Read second file 'dug.pktimes'
st2 = []
with open(scrdir/'dug.pktimes', 'r') as file2:
    for line in file2:
        hr, mm, sec = map(float, line.split()[2:5])
        atime = hr*3600+mm*60+sec
        ttime = atime - otime
        st2.append(ttime)
ns2 = len(st2)
```

```python
# Compute t2 and u2
t2, u2 = [], []
for i in range(ns2-2):
    t = st2[i+2] - st2[i]
    v = s2d / st2[i+1]
    t2.append(t); u2.append(v)

plt.plot(t2, u2)
plt.grid()
plt.show()
```

```python
# Calculate t and c
t_avg, phase = [], []
for i in range(np.min([ns1,ns2])-2):
    t = ((st1[i+2] - st1[i]) + (st2[i+2] - st2[i])) / 2
    c = deltad / ((st2[i+1] - st1[i+1]) + 0.0 * t)
    t_avg.append(t); phase.append(c)

plt.plot(t_avg, phase)
plt.grid()
plt.show()
```
